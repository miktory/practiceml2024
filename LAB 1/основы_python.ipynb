#Задача 1
Написать функцию на вход которой подается строка, состоящая из латинских букв.
Функция должна вернуть количество гласных букв (a, e, i, o, u) в этой строке. Написать тесты для кода

```python
# Your code goes here
def get_vowels(your_str):
  vowels = ['a','e','i','o','u']
  answer = []
  for i in range(0, len(your_str)):
    curr_symbol = your_str.lower()[i]
    if curr_symbol in vowels:
      answer.append(curr_symbol)
  return len(answer)

### TEST
print(get_vowels("1TEst_stri4ngFOrGet #&(+_)  5  Vo^:wels")  == 6) # Универсальная строка для теста, тест пройден
print(get_vowels("")  == 0) # Пройден
```

# Задача 2
Написать функцию на вход, которой подается строка. Функция должна вернуть true, если
каждый символ в строке встречается только 1 раз, иначе должна вернуть false. Написать тесты для кода

```python
# Your code goes here
def is_only_different_symbols(your_str):
  sym_count = {}
  for i in range(0, len(your_str)):
    curr_symbol = your_str.lower()[i]
    if curr_symbol not in sym_count:
      sym_count[curr_symbol] = 1
    else:
      sym_count[curr_symbol] += 1
  return all(value == 1 for value in sym_count.values())


### TEST
print(is_only_different_symbols("abc123") == True) # Пройден
print(is_only_different_symbols("aAbc123") == False) # Пройден
print(is_only_different_symbols("1231") == False) # Пройден
print(is_only_different_symbols("@#$%^&*aBCd") == True) # Пройден
print(is_only_different_symbols("@#$@") == False) # Пройден
print(is_only_different_symbols("") == True) # Пройден
```

# Задача 3
Написать функцию, которая принимает положительное число и возвращает количество
бит равных 1 в этом числе. Написать тесты для кода

```python
# Your code goes here

def count_positive_bits(n):
    count = 0
    if n < 0:
        raise ValueError("Число не положительное")
    while n: # цикл продолжается, пока n не станет равным нулю
        count += n & 1  # Увеличиваем счетчик, если последний бит равен 1
        n = n >> 1         # Сдвигаем число вправо на 1 бит
    return count

### TEST
print(count_positive_bits(1) == 1) # Пройден
print(count_positive_bits(2649) == 6) # Пройден. 2649 в двоичной будет 101001011001
print(count_positive_bits(0) == 0) # Пройден
try:                               # Пройден (Тест на Exception при отрицательном числе)
  count_positive_bits(-1234)
  print(False)
except ValueError:
  print(True)
```

# Задача 4
Написать функцию, которая принимает положительное число. Функция должна вернуть то,
сколько раз необходимо перемножать цифры числа или результат перемножения, чтобы
получилось число состоящее из одной цифры.
Например, для входного числа:
· 39 функция должна вернуть 3, так как 3*9=27 => 2*7=14 => 1*4=4
· 4 функция должна вернуть 0, так как число уже состоит из одной цифры
· 999 функция должна вернуть 4, так как 9*9*9=729 => 7*2*9=126 => 1*2*6=12 =>
1*2=2. Написать тесты для кода

```python
def magic(n):
  if n < 0:
    raise ValueError("Число не положительное")
  count = 0
  while len(str(n)) > 1:
    multiply_result = 1
    for i in range(0, len(str(n))):
      multiply_result *= int(str(n)[i])
    count += 1
    n = multiply_result
  return count

### TEST
print(magic(39) == 3) # Пройден
print(magic(4) == 0) # Пройден
print(magic(999) == 4) # Пройден
print(magic(0) == 0) # Пройден
try:                               # Пройден (Тест на Exception при отрицательном числе)
  magic(-1)
  print(False)
except ValueError:
  print(True)
```

# Задача 5
Написать функцию, которая принимает два целочисленных вектора одинаковой длины и
возвращает среднеквадратическое отклонение двух векторов. Написать тесты для кода

```python
# Речь про среднеквадратичную ошибку (MSE)?
def mse(pred, true):
  if len(pred) != len(true):
    raise ValueError("У векторов разная длина.")
  mse = sum((y - y_pred) ** 2 for y, y_pred in zip(true, pred)) / len(true)
  return mse

print(mse([1,2,3,4,5], [3,5,6,7,7]) == 7.0) # Пройден
print(round(mse([2,1,3], [0,1,2]),3) == 1.667) # Пройден
try:                               # Пройден (Тест на Exception при разной длине векторов)
  mse([0,1], [1,2,3,4])
  print(False)
except ValueError:
  print(True)
```

# Задача 6
Написать функцию, принимающая целое положительное число. Функция должна вернуть
строку вида “(n1**p1)(n2**p2)…(nk**pk)” представляющая разложение числа на простые
множители (если pi == 1, то выводить только ni).
Например, для числа 86240 функция должна вернуть “(2**5)(5)(7**2)(11)”. Написать тесты для кода

```python
def to_prime_multipliers(n):
  if n % 1 != 0:
    raise ValueError("Число должно быть целым")
  if n == 1:
    raise ValueError("Число должно быть больше единицы.")
  if n <= 0:
        raise ValueError("Число должно быть положительным целым!")

  factors = {}
  divisor = 2

    # Находим простые множители
  while n > 1:
      while n % divisor == 0:
          if divisor in factors:
              factors[divisor] += 1
          else:
              factors[divisor] = 1
          n //= divisor
      divisor += 1

  result = ""
  # Генерируем строку в нужном формате
  for factor, count in factors.items():
    if count == 1:
      result += ''.join(f"({factor})")
    else:
      result += ''.join(f"({factor}**{count})")
  return result

print(to_prime_multipliers(86240) == "(2**5)(5)(7**2)(11)") # Пройден
print(to_prime_multipliers(1000) == "(2**3)(5**3)") # Пройден
print(to_prime_multipliers(15) == "(3)(5)") # Пройден
print(to_prime_multipliers(2) == "(2)") # Пройден
try:                               # Пройден (Тест на Exception при единице)
  to_prime_multipliers(1)
  print(False)
except ValueError:
  print(True)

try: # Пройден (Тест на Exception при отрицательном числе (либо нуле))
  to_prime_multipliers(-1)
  print(False)
except ValueError:
  print(True)

try: # Пройден (Тест на Exception при дробном числе)
  to_prime_multipliers(5.5)
  print(False)
except ValueError:
  print(True)
```

# Задача 7
Написать функцию, принимающая целое число n, задающее количество кубиков. Функция
должна определить, можно ли из данного кол-ва кубиков построить пирамиду, то есть
можно ли представить число n как 1^2+2^2+3^2+…+k^2. Если можно, то функция должна
вернуть k, иначе строку “It is impossible”. Написать тесты для кода

```python
def pyramid(number):
  if number % 1 != 0:
    raise ValueError("Число должно быть целым")

  result = 0
  k = 0
  while result < number:
    k += 1
    result += (k * k)

  if number < 1:
    return "It is impossible"
  if result == number:
    return k
  else:
    return "It is impossible"

print(pyramid(-1) == "It is impossible")  # Пройден
print(pyramid(0) == "It is impossible")   # Пройден
print(pyramid(1) == 1)                    # Пройден
print(pyramid(3) == "It is impossible")   # Пройден
print(pyramid(5) == 2)                    # Пройден
print(pyramid(10) == "It is impossible")  # Пройден
print(pyramid(14) == 3)                   # Пройден
print(pyramid(55) == 5)                   # Пройден
try:                               # Пройден (Тест на Exception при дробном числе)
  pyramid(1.5)
  print(False)
except ValueError:
  print(True)
```

# Задача 8
Функция принимает на вход положительное число и определяет является ли оно сбалансированным, т.е. сумма цифр до средних равна сумме цифр после. Средними в случае нечетного числа цифр считать одну цифру, в случае четного - две средних. Написать тесты для кода

```python
import math

def is_balanced(number):
  if number < 0:
    raise ValueError("Число должно быть положительным.")
  skip_indexes = []
  num_str = str(number)
  skip_indexes.append(math.ceil(len(num_str) / 2) - 1)
  if len(num_str) % 2 == 0:
    skip_indexes.append(math.ceil(len(num_str) / 2))
  left_sum = 0
  right_sum = 0
  for i in range(0, len(num_str)):
    if all(i > x for x in skip_indexes):
      right_sum += int(num_str[i])
    elif all(i < x for x in skip_indexes):
      left_sum += int(num_str[i])
  return left_sum == right_sum

print(is_balanced(5) == True)      # Пройден, 0 0
print(is_balanced(24) == True)     # Пройден, 0 0
print(is_balanced(132) == False)   # Пройден, 1 2
print(is_balanced(404) == True)    # Пройден, 4 4
print(is_balanced(5135) == True)   # Пройден, 5 5
print(is_balanced(5134) == False)  # Пройден, 5 4
print(is_balanced(51315) == True)  # Пройден, 6 6

try:                               # Пройден (Тест на Exception при отрицательном числе)
  is_balanced(-1)
  print(False)
except ValueError:
  print(True)
```
